1\. File `"bank_customers.csv"` that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` describes bank clinets. It contains three columns: customer age (`CustAge`), period of relationship with the bank in months (`Months`), and credit limit (`CredLim`). Plot a histogram for the column `Month` and compute its mean, median and the mode. Show these values on the histogram. Make a second copy of the histogram, compute quartiles and show them on the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import requests


def calculate_mode(arr):
    unique_values, counts = np.unique(arr, return_counts=True)
    max_count_index = np.argmax(counts)
    mode = unique_values[max_count_index]
    return mode


filename = 'bank_customers.csv'
path = 'https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/'
data = requests.get(path + filename)
assert data.status_code == 200
text = data.text.split('\n')[1:-1]
months = np.arange(0)
for string in text:
  months = np.append(months, int(string.split(',')[1]))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(months, int(len(months) / 250), color='#85004B')
ax.set_xlabel('Months', fontsize=10)
ax.set_ylabel('Number of registrations', fontsize=10)
ax.set(xlim=[np.min(months), np.max(months)])
ax.set_title(f'Mean: {np.round(np.mean(months))}, median: {np.median(months)}, mode: {calculate_mode(months)}')
plt.show()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(months, int(len(months) / 250), color='#E667AF')
ax.set_xlabel('Months', fontsize=10)
ax.set_ylabel('Number of registrations', fontsize=10)
ax.set(xlim=[np.min(months), np.max(months)])
ax.axvline(np.quantile(months, [0.25, 0.5, 0.75])[0], label=f'Q1 at {np.quantile(months, [0.25, 0.5, 0.75])[0]}', color='black')
ax.axvline(np.quantile(months, [0.25, 0.5, 0.75])[1], label=f'Q2 at {np.quantile(months, [0.25, 0.5, 0.75])[1]}', color='black')
ax.axvline(np.quantile(months, [0.25, 0.5, 0.75])[2], label=f'Q3 at {np.quantile(months, [0.25, 0.5, 0.75])[2]}', color='black')
ax.legend()
plt.show()
```

1\. File `"bank_customers.csv"` that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` describes bank clinets. It contains three columns: customer age (`CustAge`), period of relationship with the bank in months (`Months`), and credit limit (`CredLim`). Plot a histogram for the column `CustAge` and compute its standard deviation and the interquartile range. Show them on the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import requests

filename = 'bank_customers.csv'
path = 'https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/'
data = requests.get(path + filename)
assert data.status_code == 200
text = data.text.split('\n')[1:-1]
age = np.arange(0)
for string in text:
  age = np.append(age, int(string.split(',')[0]))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(age, int(len(age) / 250), color='#A64B00')
ax.set_xlabel('Customer age', fontsize=10)
ax.set_ylabel('Number of registrations', fontsize=10)
ax.set(xlim=[np.min(age), np.max(age)])
sigma = np.std(age)
q25 = np.quantile(age, 0.25)
q75 = np.quantile(age, 0.75)
iqr = q75 - q25
ax.axvline(np.mean(age) + sigma, color='black', label=f' = {np.round(sigma)}')
ax.axvline(np.mean(age) - sigma, color='black')
ax.axvline(q25, color='black', label=f'IQR = {iqr}', linestyle='dashed')
ax.axvline(q75, color='black', linestyle='dashed')
ax.legend()
plt.show()
```

2\. Download the dataset from the previous exercise and compute correlation coefficients for each pair of columns.

```python
import numpy as np
import matplotlib.pyplot as plt
import requests


def correlation_coefficient(arr1, arr2):
    mean_arr1 = np.mean(arr1)
    mean_arr2 = np.mean(arr2)

    numerator = np.sum((arr1 - mean_arr1) * (arr2 - mean_arr2))
    denominator = np.sqrt(np.sum((arr1 - mean_arr1)**2) * np.sum((arr2 - mean_arr2)**2))

    correlation_coeff = numerator / denominator

    return correlation_coeff


filename = 'bank_customers.csv'
path = 'https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/'
data = requests.get(path + filename)
assert data.status_code == 200
text = data.text.split('\n')[1:-1]
age = np.arange(0)
months = np.arange(0)
credlim = np.arange(0)
for string in text:
  age = np.append(age, int(string.split(',')[0]))
  months = np.append(months, int(string.split(',')[1]))
  credlim = np.append(credlim, float(string.split(',')[2]))
print(f'Corellation coefficient for age and months: {np.round(correlation_coefficient(age, months), 3)}')
print(f'Corellation coefficient for age and credit limit: {np.round(correlation_coefficient(age, credlim), 3)}')
print(f'Corellation coefficient for months and credit limit: {np.round(correlation_coefficient(months, credlim), 3)}')
```
