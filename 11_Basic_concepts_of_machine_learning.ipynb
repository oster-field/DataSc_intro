*Модель* — это в той или иной мере **упрощенное** представление реального процесса, системы или явления. Модель фиксирует основные взаимосвязи и закономерности в данных. Это нужно для прогнозирования. Модели могут быть разными по своей природе, например, математическими.

*Модель машинного обучения* - это алгоритм, который обучается на данных для выполнения определенной задачи, такой как классификация, регрессия или кластеризация, без явного программирования. По сути - это математическое представление входных данных, которое используется как раз для прогнозирования.



1.   **Контролируемое обучение**. Это тип машинного обучения, при котором модель обучается на основе размеченных человеком данных. Обучающий "набор" данных представлен с соответствующей выходной меткой или целевым значением. Цель состоит в том, чтобы научить модель предсказывать или выводить правильные выходные значения на основе входных данных. Например, задачи классификации являются классическими примерами такого обучения. Так же, насколько я знаю, дерево решений - тоже контролируемое обучение.
2.   **Неконтролируемое обучение**. Наоборот производится без маркировки данных человеком. Цель состоит в том, чтобы модель самостоятельно обнаруживала скрытые закономерности, структуры или группы в данных. Примером является алгоритм PCA из секции 10_Dimensional_Reduction.
3.   **Обучение с подкреплением**. Включает в себя обучение модели принимать последовательность решений в некоторой среде для достижения конкретной цели. Модель учится методом проб и ошибок, получая обратную связь в виде вознаграждений или наказаний в зависимости от своих действий. Алгоритмы обучения с подкреплением направлены на максимизацию общего вознаграждения с течением времени. *   **Overfitting** - модель машинного обучения слишком хорошо изучила обучающие данные, до той степени, когда начинает воспроизводить даже шум, который иногда встречается в обучающих данных, как если бы этот шум был важной закономерностью. В результате модель хорошо работает с данными, которые максимально похожи на обучающие, но "сыпется" на других данных, для которых, в целом, она и предназначена.
*   **Underfitting** - обратная ситуация, модель после обучения плохо работает как с обучающими данными, так и со всеми остальными. В этом случае модель слишком проста и не может выявить закономерности данных, поскольку имеет мало параметров.
*   Разбиение набора данных на обучающие, проверочные (валидационные) и тестовые наборы необходимо для непосредственного обучения, а затем и оценки работы модели для предотвращения переобучения и недостаточного обучения.



Способы борьбы с переобучением:


1.   Regularization. Ограничение среднеквадратического отклонения значений параметров.
2.   Dropout. Случайным образом деактивируется часть параметров в процессе обучения, что не позволяет модели слишком сильно концентрироваться на каких-либо конкретных параметрах.
3.   Cross-validation. Это процедура, используемая в машинном обучении для оценки модели на независимых данных. Основная идея кросс-валидации заключается в том, что вместо разделения обучающего набора данных на обучающий и валидационный набор один раз, данные разбиваются на несколько (обычно k) независимых неперекрывающихся частей. Каждая часть содержит примерно одинаковое количество данных. На каждой итерации одна из частей используется в качестве тестового набора данных, а остальные части - в качестве обучающих данных. Модель обучается на обучающих данных и оценивается на тестовых данных. Этот процесс повторяется k раз, так что каждая часть данных используется для валидации ровно один раз.

Способы борьбы с недостаточным обучением является обратной задачей, необходимо усложнение модели, увеличение объема входных данных, увеличение среднеквадратического отклонения значений параметров.








Для оценки эффективности регрессионных моделей используется несколько метрик:

1.   Mean Squared Error (MSE): Она измеряет среднеквадратическую разницу между прогнозируемыми и прогнозируемыми значениями.
2.   Root Mean Squared Error (RMS): Это квадратный корень из MSE.
3.   Mean Absolute Error (MAE): Она вычисляет среднее значение абсолютных разностей между фактическими и прогнозируемыми значениями.
4.   R-squared: оценивает долю дисперсии зависимой переменной, которую можно предсказать на основе независимых переменных. Она варьируется от 0 до 1.
5.   Mean Absolute Percentage Error (MAPE): измеряет среднее значение абсолютных значений разности между фактическими и прогнозируемыми значениями (в процентах).
6.   Root Mean Squared Logarithmic Error (RMSLE): нужна если необходима измерить экспоненциальный рост параметра, считается аналогичным образом с RMS, но использует натуральный логарифм.
