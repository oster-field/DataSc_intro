1\. Create a plot with graphs of two functions: $y=e^{-x^2}$ and $y=3x^2 e^{-x^2}$. Add x and y axes labels and the legend. Turn on the grid. Customize curve colors and styles.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.01)
y1 = np.exp(-x**2)
y2 = 3 * x**2 * y1

fig, ax = plt.subplots()
ax.plot(x, y1, c='#29524F', linestyle='dashed', label='exp(-x²)')
ax.plot(x, y2, c='#855F42', linestyle='dashdot', label='3x²exp(-x²)')
ax.set_xlabel('x', fontsize=10)
ax.set_ylabel('y', fontsize=10)
ax.legend(fontsize=10)
ax.grid()
plt.show()
```

2\. File `"apple_vs_cabot.csv"` that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` contains stock prices of "Apple Inc." and "Cabot Oil & Gas" recorded  from 08 Feb 2013 till 07 Feb 2018. For simplicity instead of the actual date a day number within this range is stored in the file. Using an example above read this file and create a line plot demonstrating variation of theses two prices.

```python
import numpy as np
import matplotlib.pyplot as plt
import requests

filename = 'apple_vs_cabot.csv'
path = 'https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/'
data = requests.get(path + filename)
assert data.status_code == 200
text = data.text
valuable_text = text.split('\n')[1:-1]
apple = np.arange(0)
cabot = np.arange(0)
for string in valuable_text:
  apple = np.append(apple, float(string.split(',')[1]))
  cabot = np.append(cabot, float(string.split(',')[2]))

fig, ax = plt.subplots()
ax.plot(apple, c='#078507', label='Apple')
ax.plot(cabot, c='#A60909', label='Cabot')
ax.set_xlabel('day', fontsize=10)
ax.set_ylabel('price', fontsize=10)
ax.legend(fontsize=10)
plt.show()
```

3\. Above is the dataset describing world total energy supply recorded at 2015. Create a bar plot for it. Write text labels with values on each bar. Add x and y axes labels.

```python
dat = [3842742,2928795,670172,334851,203821,1271235,4328233]
ticks = ["Coal", "Natural gas", "Nuclear", "Hydro", "Wind, solar, etc", "Biofuels and waste", "Oil"]

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Total energy supply (TES) by source, World 2015')
chart = ax.bar(range(len(dat)), dat, color='#4A096F', tick_label=ticks)
ax.bar_label(chart, fmt='{:,.0f}')
ax.set_xlabel('Type of energy supply', fontsize=10)
ax.set_ylabel('Thousand tons of oil equivalent', fontsize=10)
ax.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
plt.xticks(rotation=90)
plt.show()
```

4\. File `"moscow_temp.csv"` that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` contains daily records of average temperature in Moscow for many years. Read this file and create a histogram showing how often certain temperatures are registered. The data are represented in Fahrenheit. Before plotting convert them into  Celsius. Add appropriate axes labels.

```python
import numpy as np
import matplotlib.pyplot as plt
import requests

filename = 'moscow_temp.csv'
path = 'https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/'
data = requests.get(path + filename)
assert data.status_code == 200
text = data.text
valuable_text = text.split('\n')[1:-1]
t = np.arange(0)
for string in valuable_text:
  celsius = (float(string.split(',')[3]) - 32) / 1.8
  if celsius < -70:
    print(f"It was quite cold in Moscow {string.split(',')[0]}.{string.split(',')[1]}.{string.split(',')[2]}: {string.split(',')[3]} °C")
  else:
    t = np.append(t, celsius)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(t, bins=int(len(t)/50), color='#771BAA')
ax.set_xlabel('Temperature in Celsius', fontsize=10)
ax.set_ylabel('Number of registrations', fontsize=10)
plt.show()
```
